```python
import pandas as pd
import glob
```

```python
files = glob.glob("*.xlsx")
print(files)

dfs = [pd.read_excel(f) for f in files]

combined_df = pd.concat(dfs, ignore_index=True)
```

```python
len(combined_df)
```

```python
combined_df.to_csv("Combined_Tiktok_Comments.csv", index=False, encoding="utf-8")
```
